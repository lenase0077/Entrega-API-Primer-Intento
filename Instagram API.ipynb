## Algo raro que hice con bucle for

```python
artists = ['9695086', '58585', '119', '13', '11549549']

url = "https://deezerdevs-deezer.p.rapidapi.com/artist/{artist_id}"

headers = {
	"X-RapidAPI-Key": "e4a9dea8cfmsh2fcab100d15a52bp1829cbjsna7392e81e028",
	"X-RapidAPI-Host": "deezerdevs-deezer.p.rapidapi.com"
}

artist_data = []

for artist in artists:
    response = requests.request("GET", url.format(artist_id=artist), headers=headers)
    data = response.json()
    artist_data.append(data)

print(artist_data)

df = pd.DataFrame(artist_data)
df.head()
```
